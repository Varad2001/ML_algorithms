### Gradient descent algorithm from scratch

##### What does it do? :
It finds out the local minima, i.e. , the input value at which the given output(or function) value will be minimum.
##### How does it work?
A random input value is picked and gradient at that value is calculated. We take the partial derivatives of the given function and use them to calculate next point of input. We use gradients because they give us the direction of the steepest descent,i.e., where the value will change by the maximum amount if moved in the given direction.

#### Steps:
1. Pick a random input value.
2. Calculate gradient at that point.
3. Now based on the gradient, change the input parameter by multiplying it by a small value which is called learning_rate.
4. Calculate the new point for the input.
5. Repeat steps 2,3,4 until iterations become very large or the difference between the two points become very less.

#### Gradient descent for a function of two independent variables
For a given function, the algorithm will return the minima.

```python
## define a function in two variables
z = lambda x,y : (y+3*x)**2
```

```python
## displaying graph
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-3,3,0.1)
y = np.arange(-2,4,0.1)
X, Y = np.meshgrid(x,y)
Z = (3*X+Y)**2
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
## define the gradient for the given function
## partial derivative w.r.t x at the point (x,y)
derivative_x = lambda x,y : 6*(3*x+y)
## partial derivative w.r.t y at the point (x,y)
derivative_y = lambda x,y : 2*(3*x+y)
```

```python
from matplotlib.pyplot import plot
tx = map(derivative_x,x,y)
ty = map(derivative_y,x,y)
der_x = list(tx)
der_y = list(ty)
plot(x,der_x)
plot(y,der_y)
## obviously a straight lines as derivatives are linear functions 
```

```python
## pick a random point
pt = [4,5]
print(z(pt[0],pt[1]))
print(derivative_x(pt[0],pt[1]))
print(derivative_y(pt[0],pt[1]))
```

```python
## define learning_rate and max_iterations
rate = 0.01
max_iter = 100
```

```python
## iterate until the slopes are near zero, i.e., der_x & der_y = 0.01 within 100 iterations
pt = [4,5]
i = 0   ## count iterations
print("Iteration : ",i)
print("Current point:",pt[0],pt[1])
print("Function value:",z(pt[0],pt[1]))
print()
while i<max_iter and (derivative_x(pt[0],pt[1]) > 0.000001 and derivative_y(pt[0],pt[1]) > 0.000001):
    pt[0]=pt[0] - rate*derivative_x(pt[0],pt[1])
    pt[1] -= rate*derivative_y(pt[0],pt[1])
    i=i+1
    print("Iteration : ",i)
    print("Current point:",pt[0],pt[1])
    print("Function value:",z(pt[0],pt[1]))
    print()
```

```python
print("The minima is:(%0.4f , %0.4f)" %(pt[0],pt[1]))
print("And the minimum value is:" ,z(pt[0],pt[1]))
```
